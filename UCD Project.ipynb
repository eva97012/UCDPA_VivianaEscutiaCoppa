# Merging/Joing tables: Part 1

```python
#Joining dataframes "lifestyle" with "country" to achieve granularity by country

LS_Country = lifestyle.merge(country, left_on= 'City', right_on= 'city', how="left")

#reviewing shape after the join. The join resulted in some of the cities being duplicated.Now the total of cities is 89 instead of 44
LS_Country.shape
```

```python
# "Amsterdam" is one of the cities duplicated joined to two countries: Netherlands and United States.

LS_Country.head(2)
```

```python
# Many cities are showing under Country United States which is incorrect.

LS_Country.loc[LS_Country.country == 'United States', :]
```

# Dropping Duplicates

```python
# Keeping the first of the duplicated values will eliminate the city/country incorrect joins.

LS_Country.drop_duplicates(['City'], keep= 'first', inplace=True)
```

```python
LS_Country.shape
```

```python
#afte deleting duplicated values "Amsterdam" is showing with the correct Country
LS_Country.head(2)
```

```python
# United States is related to the correct cities.

LS_Country.loc[LS_Country.country == 'United States', :]
```

# Merging/Joing tables: Part 2


```python
#Joining dataframes "lifestyle/Country" with "continents" to achieve granularity by continent

LS_Continent = LS_Country.merge(continents, left_on= 'iso3', right_on= 'code_3',how="left")
LS_Continent
```

# Defining a Function

```python
#BMI Calculator

name1 = "James"
height_m1= 2
weight_kg1= 95

name2 = "John"
height_m2= 1.75
weight_kg2= 80

name3 = "Mark"
height_m3= 1.70
weight_kg3= 150
```

```python
def bmi_calculator(name, height_m, weight_kg):
    bmi = weight_kg/ (height_m **2)
    print("bmi: ")
    print(bmi)
    if bmi < 25:
        return name + " is not overweight"
    else:
        return name+ " is overweight"
```

```python
result1 = bmi_calculator(name1, height_m1, weight_kg1)
result2 = bmi_calculator(name2, height_m2, weight_kg2)
result3 = bmi_calculator(name3, height_m3, weight_kg3)
```

```python
print(result1)
print(result2)
print(result3)
```

# Visualisations 

```python
from matplotlib import pyplot as plt
import pandas as pd

plt.style.use('bmh')
df= pd.read_csv("healthy_lifestyle_city_2021.csv")

specified = df.loc[df['Rank'] <= 5]
x = specified['City']
y= specified['Rank']


plt.xlabel('City', fontsize=16)
plt.ylabel('Rank', fontsize=16)
plt.title("Top 5 Cities with Healthy Lifestyles")

plt.bar(x,y,color=['purple','g','y','b','orange'])

plt.show()

```

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns

df= LS_Continent

sns.set_style("whitegrid")
custom_palette = ["red","green","orange","blue","yellow","purple"]
sns.set_palette(custom_palette)

g= sns.relplot(x="Outdoor activities(City)", y= "Rank", data=df, kind="scatter", size="continent", hue= "continent")
plt.xticks(rotation=90)
g.fig.suptitle("Correlation between Outdoor activities spent in hours and Level of Healthy Lifestyle",y=1.03)
plt.show()
```
